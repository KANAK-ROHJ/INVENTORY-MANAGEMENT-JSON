```python
import json

#importing time module to record the time of transaction using ctime() function
import time

#reading inventory data from Record.json file
fd=open("Record.json","r")
js=fd.read()
fd.close()

#Converting json file data to Dictionary--(record)
record=json.loads(js)

#------------------------------
#Displaying items
print("list of items".upper())
print("-"*60)
print("ID  DETAILS")
for key,value in record.items():
    print(key," ",value)
print("-"*60)
print("\n\n")

#--------------------------------
#user details
print("-"*60)
u_name=   input("Enter your name                    : ")
u_phno=   input("Enter your phone no.               : ")
u_mail=   input("Enter your Email id                : ")

#dictionary to record items(key) purchased by user and quantities(value)
#prod=== purchased items dictionary
prod={}


print("-"*60)

#-----------------------------
#taking products details from the user 
while True:
    
    up_id=input("Enter product id                   : ")
    
    #if the id does not exist, inform the user about wrong input and execute the loop again
    if up_id not in record:
        print("*"*60)
        
        print("Wrong product id")
        
        print("*"*60)
        
        #skip rest of the code
        continue
        
    up_qn=int(input("Enter product quantity             : "))
    
    #adding the id and quantity to the dictionary as key-value pair
    prod[up_id]=up_qn
    
    #if user wants more than available quantity of a product
    if(up_qn>record[up_id]["Qn"]):
        
        #available quantity of the product
        qn=record[up_id]["Qn"]
        
        nm=record[up_id]["Name"]
        print("*"*60)
        
        #inform the user about the available quantity
        print(f"Only {qn} {nm} available")
        
        #asking the user if he wants available quantity or not
        inp=input(f"Want to buy {qn} {nm} ? Y/N    : ")
        print("*"*60)
        
        #if user wants available quantity, then update the "prod" dictionary with available quantity
        if inp!="N" and inp!="n":
            prod[up_id]=qn
        
        #if user doesn't want product with available quantity then delete this entru from purchased items dictionary----"prod"
        else:
            del prod[up_id]
    
    #asking if the user wants to purchase more items
    more =input("Want to purchase more items? Y/N   : ")    
    if(more=="N" or more=="n"):
        break
print("-"*60)
print("\n\n")

#------------------------------
#generating bill
print("BILL")

#variable for recording total amount payable by user
total=0
print("-"*60)

# going through the purchased items dict--"prod"
for key in prod.keys():
    pname=record[key]["Name"]
    pprice=record[key]["Price"]
    pqn=prod[key]
    print("Product Name              : ",pname)
    print("Price                     : ",pprice)
    print("Quantity                  : ",pqn)
    
    #total=total+ (price*quantity)
    total+=pprice*pqn
    
print("-"*60)
print("Total Amount              : ",total)
print("-"*60)
print()

#-------------------------------------
#updating inventory(Record.json) file
for key,value in prod.items():
    
    #updating quantity of purchased products in record dictionary
    record[key]["Qn"]-=value
    
#converting record dictionary to json string
js=json.dumps(record)

#updating the Record file with new js string
fd=open("Record.json","w")
fd.write(js)
fd.close()

#--------------------------------------
#updating transaction file

#readiing the data of transaction file to get the length of trans dictionary
fd=open("Transaction.json","r")
js=fd.read()

#convert json string to trans dictionary
if js:
    trans=json.loads(js)
    
#if transaction file is empty then create a empty trans dictionary
else:
    trans={}
    
#transation id-- length of dictionary +1
#in this way we can give unique id to each transaction\
tid=len(trans)+1
fd.close()

#opening file in write mode 
fd=open("Transaction.json","w")

#adding new entry to  trans dictionary
trans[tid]={"Time":time.ctime(),"U_Name":u_name ,"U_Phone":u_phno,"U_Email":u_mail,"Items":prod,"Amount":total}

#dictionary to json string
njs=json.dumps(trans)
#overwrite complete trans dictionary to transaction file
fd.write(njs)
#close file
fd.close()


#END OF PROJECT-------------------------------------------
#---------------------------------------------------------
```
