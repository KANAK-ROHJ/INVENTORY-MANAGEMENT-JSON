# Inventory Management (with files)
## Product Details
1. Product ID
2. Product Name
3. Price
4. Quantity

```python

#Taking user details
uname=input("Enter your name : ")
uphone=input("Enter your phone no. : ")
uemail=input("Enter your Email id : ")

#dictionary recording the id(key) and quantity(value) of items which user wants to buy
prod={}

#Taking items details 

while True:
    pi=input("Enter Product ID:")
    qn=int(input("Enter Product Quantity:"))
    
    #Adding item id and quantity in the dictionary
    prod[pi]=qn
    nxt=input("Want to purchase more items?? Y/N: ")
    
    #if user doesn't want to buy more items
    if nxt=="N" or nxt=="n":
        break


#Reading the Inventory file
fd=open("Inventory","r")
txt=fd.read()
fd.close()

#list of products in Inventory
products=txt.split("\n")

# variable for storing total amount payable by user
total=0
print("-"*60)

#new -products list-(with updated quantity available) to be stored in Inventory file after purchase
updated_products_list=[]

#going through the products list
for product in products:
    
    #list contaning details of the particular product
    p_details=product.split(',')
    id=p_details[0]
    
    #checking if user wants to buy this product
    if id in prod:
        quan=prod[p_details[0]]
        name=p_details[1]
        price=p_details[2]
        
        #bill
        print("Product ID                  : ",id)
        print("Product Name                : ",name)
        print("Price                       : ",price)
        print("Quantity                    : ",quan)
        print()
        
        #adding total cost of this product(price*quantity) to total amount
        total+= int(price) * quan
        
        #updating the quantity of the purchased product
        p_details[3]=str(int(p_details[3]) - quan)
    
    #adding the product details(with updated quantity) to updated products list
    updated_products_list.append(",".join(p_details))
    
print("-"*60)
#total amount to be printed on bill
print("Total Amount                : ",total)
print("-"*60)

#converting the updated product list from list to string 
#final text to be written on Inventory file
final_txt="\n".join(updated_products_list)

#updating Inventory file
fd=open("Inventory","w")
fd.write(final_txt)
fd.close()
```
